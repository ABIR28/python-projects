Join two tables and read them in a dataframe using read_sql_query

```python

query = '''
 SELECT students.student_id,students.name, students.phone, courses.subject
 FROM students INNER JOIN courses ON students.student_id=courses.id
'''

df = pd.read_sql_query(query,engine)
df
```

```python

query = '''
 SELECT students.student_id, students.phone, courses.id,courses.name, courses.subject
 FROM students INNER JOIN courses ON students.student_id=courses.id
'''
df = pd.read_sql_query(query,engine)
df
```

```python
query = '''
 SELECT students.student_id,students.name, students.phone, courses.subject
 FROM students left JOIN courses ON students.student_id=courses.id
'''
df = pd.read_sql_query(query,engine)  #left join
df
```

```python
query = '''
 SELECT students.name, students.phone, courses.name, courses.subject
 FROM students right JOIN courses ON students.student_id=courses.id
'''
df = pd.read_sql_query(query,engine)  #left join
df
```

Write to mysql database using to_sql


```python
df1 = pd.DataFrame({
    'id': [15,16,17,18],
    'name': ['emily','derek','diana','avon'],
    'subject' :['machine learning', 'deep learning','bigdata','reinforcement learning']
})

df1
```

```python
df1.to_sql(
    name='courses', # database table name
    con=engine, #connection
    if_exists ='replace', #append with existing records,by deafult fail (append,fail,replace)
    index=False   #do not write the index
)
```

Execution of other queries like, create,insert,delete

```python
import mysql.connector
```

```python
mydb=mysql.connector.connect(host="localhost",user="root",password ="1234",database="ailabs")
```

```python
cur = mydb.cursor()
```

CREATE

```python
cur.execute("create table customer(id int(10) PRIMARY KEY,name varchar(40),amount int(100))")
mydb.commit()
```

DELETE

```python
cur.execute("drop table customer;")
mydb.commit()
```

INSERT

```python
cur.execute("insert into customer values(03,'krishna',800)") #insert 3 rows
mydb.commit()
```

READ

```python
cur.execute("select * from customer")
result = cur
```

```python
for i in result:
    print(i)
```

```python
cur.execute("select * from customer where amount between 600 and 700")
result = cur
for i in result:
    print(i)
```
